```python
import pandas as pd
try:
    input1=float(input("Stream 1 is "))
    print(f"Stream 1 is {input1:.2f}kg")
    input2=float(input("Stream 2 is "))
    print(f"Stream 2 is {input2:.2f}kg")
    Total_flow=input1+input2
    print(f'Output is {Total_flow}kg')
    if abs((input1+input2)-Total_flow)<0.00001:
     print("Mass Balance Satisfactory")
    else:
     print("Mass Balance Not Satisfactory")
    data={
        "Component": ["Stream 1","Stream 2","Total_flow"],
        "Flowrate (kg/h)": [input1,input2,Total_flow]
    }
    df=pd.DataFrame(data)
    print(df)

except ValueError:
    print("Invalid input. Please enter a number.")
```
